```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# plot bar graph of category count
df = pd.read_csv('glycemic_index_data.csv')
df['category'].value_counts().plot(kind='bar')
```

```python
plt.figure(figsize=(9, 6))
sns.scatterplot(data=df, x='glycemic_load', y='glycemic_index', hue='category')
# Adjusting plot labels and title
plt.title('Scatter Plot of Glycemic Load vs. Glycemic Index')
plt.xlabel('Glycemic Load')
plt.ylabel('Glycemic Index')
plt.legend(title='Category', bbox_to_anchor=(1.01, 1), loc='upper left')

# Show plot
plt.tight_layout()
plt.show()
```
